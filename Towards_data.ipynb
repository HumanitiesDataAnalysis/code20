# Pandas

Pandas is the library to use for data analysis in python. It's sometime ugly, sometimes crazy fast, sometimes crazy slow... But there's no real alternative. If you want to work with a metropolitan museum CSV, for example, you'll have to read it in thusly:

```
import pandas as pd 

met = pd.read_csv('https://media.githubusercontent.com/media/metmuseum/openaccess/master/MetObjects.csv')
```

```
data = met.groupby(["Department", "Object Date"])['Object ID'].count().reset_index()

# Can you see what's wrong with this regular expression?

clean_year = data[data["Object Date"].str.match(r"^[0-9]{3,4}$")].copy()
clean_year['year'] = clean_year['Object Date']
```

```
import altair as alt
```

```
alt.Chart(clean_year, width=800)\
  .mark_line()\
  .encode(x="year:Q", color="Department", y="Object ID")
```
